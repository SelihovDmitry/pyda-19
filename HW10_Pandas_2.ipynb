## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
def classification(rating):
    if rating <= 2:
        return 'Bad'
    elif rating <= 4:
        return 'Normal' 
    elif rating > 4:
        return 'Good'
    
# Вопрос: попробовал написать так: return 'Bad' if rating <= 2 - не работает, invalid syntax, 
# нельзя в одну строку if написать в return ?
        
```

```python
df_ratings['class'] = df_ratings['rating'].apply(classification)
df_ratings.sample(10)
```

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

1 'Центр': ['москва', 'тула', 'ярославль'],

3 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
df_kw = pd.read_csv('keywords.csv')
df_kw.sample(10)
```

```python
def what_region(keyword):
    if ('москв' or 'тул' or 'ярославл') in keyword:
        return 'Центр'
    elif ('петербург' or 'псков' or 'мурманск') in keyword:
        return 'Северо-Запад'
    elif ('владивосток' or 'сахалин' or 'хабаровск') in keyword:
        return 'Дальний Восток'
    return 'undefined'

# опять же, может как-то изящнее можно чем через if, elif....
```

```python
df_kw['region'] = df_kw['keyword'].apply(what_region)
df_kw.sample(10)
```

```python
# посмотримб есть ли в определении регионов что-то кроме undefined
df_kw['region'].value_counts().head()
```

## Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
df_mov = pd.read_csv('movies.csv')
df_mov.head()
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
years = range(1950, 2011)
years
```

```python
def production_year(title):
    # проходим в цикле по всем годам и ищем строковое совпадение в переданной строке (названии фильма)
    for year in years:
        if str(year) in title:
            return year # возвращаем год если он есть в названии
    return 1900 # возвращием 1900 если года в названии нет или год вне диапазона из условия задачи
```

```python
df_mov['year'] = df_mov['title'].apply(production_year)
df_mov.sample(10)
```

В принципе видно что функция работает верно, т.е. для фильмов вне диапазона 1950 - 2010 выставляется 1900 год.

```python
df_total = df_ratings.merge(df_mov, how = 'left', left_on='movieId', right_on='movieId')
df_total.head()
```

Группируем по годам, считаем средний рейтинг и выводим по убыванию рейтинга.

```python
df_total[['year', 'rating']].groupby('year').mean('rating').sort_values(by='rating', ascending=False).reset_index().head(20)
```
