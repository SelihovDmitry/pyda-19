## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
df_log = pd.read_csv('visit_log.csv', sep=';')
df_log.head()
```

```python
df_log.loc[ (df_log.traffic_source == 'yandex') | (df_log.traffic_source == 'google'), 'source_type'] = 'organic'
df_log.loc[ ( (df_log.traffic_source == 'email') | (df_log.traffic_source == 'paid') ) 
           & (df_log.region == 'Russia'), 'source_type'] = 'ad'
df_log.loc[ (df_log.traffic_source.isin(['paid', 'email']))
           & (df_log.region != 'Russia'), 'source_type'] = 'other'
df_log.sample(10)
```

```python
df_log.loc[pd.isna(df_log.source_type), 'source_type'] = df_log.traffic_source
df_log.sample(10)
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_user_lifetime = pd.read_csv('ml-latest-small/ratings.csv')
df_user_lifetime.head()
```

```python
user_max = df_user_lifetime[['userId', 'timestamp']].groupby('userId').max().reset_index()
user_max.head()
```

```python
user_min = df_user_lifetime[['userId', 'timestamp']].groupby('userId').min().reset_index()
user_min.head()
```

```python
# количество оценок пользователей
user_count = df_user_lifetime[['userId', 'timestamp']].groupby('userId').count().reset_index()
user_count.head()
```

```python
# фильтр пользователей с оценкаим больше 100
user100 = user_count[user_count.timestamp > 100]
user100.head()
```

```python
# теперь начинаем все это соединять
total = user100.merge(user_max, on='userId', how='left').merge(user_min, on='userId', how='left')
total.head(10)
```

```python
total.rename(columns = {'timestamp_x': 'count', 'timestamp_y': 'time_out', 'timestamp': 'time_in'}, inplace=True)
total.head()
```

```python
# время жизни в днях каждого пользователя (86400 - кол-во секунд в сутках)
total['lifetime'] = round((total.time_out - total.time_in) / 86400, 2)
total.head(10)
```

```python
# Ну и ответ на поставленный вопрос - среднее время жизни пользователей со 100+ оценками
round(total['lifetime'].mean(), 2)
```

### Ответ 463.89
