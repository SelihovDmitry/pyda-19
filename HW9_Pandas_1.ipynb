## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
df1 = pd.read_csv('ratings.csv')
```

```python
df1.head()
```

```python
df1.info()
```

```python
 df2 = pd.read_csv('movies.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
# выделим в отдельны ДФ фильмы с оценкой 5.0
df1_rating_5_0 = df1[df1.rating == 5.0]
df1_rating_5_0.head()
```

```python
# Посмотрим какому фильму поставили больше всего оценок 5.0
top_mov = df1_rating_5_0['movieId'].value_counts().head()
top_mov
```

```python
# Id фильма с наибольшим количесвом оценок 5.0 будет таким
top_mov_Id = top_mov.index[0]
top_mov_Id
```

```python
# Название фильма с наибольшим количесвом оценок 5.0 будет таким
top_mov_name = df2[df2.movieId == top_mov_Id].title.values[0]
top_mov_name
```

```python
# Выведем результат
print(f'Фильм, которому поставили больше всего оценок 5.0 называется: "{top_mov_name}", \nему поставили {top_mov.values[0]} максимальные оценки.')
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df_power = pd.read_csv('power.csv')
df_power.head()
```

```python
df_power.info()
```

```python
# найдем как называется Латвия, Литва и Эстония в этом датафрейме
df_power_baltic = df_power[df_power['country'].str.contains('lat', case=False) | df_power['country'].str.contains('lit', case=False) | df_power['country'].str.contains('est', case=False)]
df_power_baltic.head()
```

```python
df_power_baltic.country.unique()
```

Видим что есть 2 "лишние страны", улучшим фильтр и оставим только нужные нам

```python
df_power_baltic = df_power[df_power['country'].str.contains('Latvia', case=False) | df_power['country'].str.contains('Lithuania', case=False) | df_power['country'].str.contains('Estonia', case=False)]
df_power_baltic.country.unique()
```

Теперь фильтр по странам балтии верный, можно идти дальше

```python
df_power_baltic[ ( df_power_baltic.quantity > 0 ) & ( df_power_baltic.category == 4 ) ].head()
```

```python
df_power_baltic[df_power_baltic.category == 4].head()
```

```python
df_power_baltic.query('quantity > 0 and category == (4,12,21) and year >= 2005 and year <= 2010').sample(10)
```

```python
# запишем этот результат в новый датафрейм
df_filter = df_power_baltic.query('quantity > 0 and category == (4,12,21) and year >= 2005 and year <= 2010')
df_filter.sample(10)
```

```python
# осталось только получить сумму
sum_cons_baltic = df_filter.quantity.sum()
sum_cons_baltic
```

#### Ответ

```python
print(f'Cуммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года равно {sum_cons_baltic}')
```
